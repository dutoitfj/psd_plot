```python
# import required modules
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.models import Range1d, PrintfTickFormatter
from bokeh.io import output_notebook

output_notebook()
```

```python
# function for generating a list of sample ids
def generate_sample_id_list(file):
    sample_ids = pd.ExcelFile(file).sheet_names
    return sample_ids

# function for reading and preparing the data from the .xlsx file
def read_data(file, sheet_name):
    df = pd.read_excel(file, sheet_name)
    df_transposed = df.transpose()
    return df_transposed

# function for generating the plots
def generate_plot(psd_data):
    # generate a semilog plot
    p = figure(
    width = 400,
    height = 400,
    x_axis_type = 'log',
    title = 'Particle Size Distribution: ' + sample_ids[i],
    toolbar_location = None
    )
    # add scatter plot
    p.scatter(
        x = psd_data.iloc[0],
        y = psd_data.iloc[1],
        color = 'black',
        alpha = 0.5
    )
    # add line plot
    p.line(
        x = psd_data.iloc[0],
        y = psd_data.iloc[1],
        color = 'black',
        alpha = 0.25
    )
    # style the plot
    p.title.text_font_size = '10pt'
    p.xaxis.axis_label = 'Particle Diameter (mm)'
    p.yaxis.axis_label = 'Cumulative Mass Percentage (%)'
    p.y_range = Range1d(start = 0, end = 100)
    p.x_range = Range1d(start = 0.001, end = 100)
    p.xaxis.formatter = PrintfTickFormatter(format = '%.2f')
    p.outline_line_color = '#A9A9A9'
    p.grid.grid_line_color = '#A9A9A9'
    p.grid.minor_grid_line_color = '#D3D3D3'
    # generate the plot
    show(p)
```

```python
# define the name of the file containing the data
file = 'PSD Data Combined.xlsx'
```

```python
# generate a list of sample ids
sample_ids = generate_sample_id_list(file)

# run a for loop on each sheet in the .xlsx file (each sample)
for i in range(len(sample_ids)):
    # generate a panads dataframe
    psd_data = read_data(file, sample_ids[i])
    # generate the particle size distribution plot
    generate_plot(psd_data)
```
